<h1><font color = 'brown' size = '6'>
<b>
Horizontal and Vertical Shift Augmentation
</b>
</font>
</h1>

<h1>
<b>
<ul>
<font color = 'brown green' size = '5'>

<li>
A shift to an image means moving all pixels of the image in one direction, such as horizontally or vertically, while keeping the image dimensions the same.
</li><br>

<li>
This means that some of the pixels will be clipped off the image and there will be a region of the image where new pixel values will have to be specified.
</li><br>

<li>
The width shift range and height shift range arguments of the <i>ImageDataGenerator</i> constructor control the amount of horizontal and vertical shift respectively.
</li><br>

<li>
These arguments can specify a floating point value that indicates the percentage (between 0 and 1) of the width or height of the image to shift.
</li>

</font>
</ul>

</b>
</h1>

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# example of horizontal shift image augmentation
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from numpy import expand_dims
import matplotlib.pyplot as plt

# load the image
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bird.jpg')

# convert to numpy array
data = img_to_array(img)

# expand dimension to one sample
samples = expand_dims(data, 0)

# create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range = [-100, 100])

# prepare iterator
it = datagen.flow(samples, batch_size = 1)

# generate samples and plot
fig = plt.figure(figsize = (20, 5))
for i in range(10):
  # define subplot
  plt.subplot(2, 5, i+1)
  # generate batch of images
  batch = it.next()
  # convert to unsigned integers for viewing
  image = batch[0].astype('uint8')
  # plot raw pixel data
  plt.imshow(image)
```

```python
# example of vertical shift image augmentation

# create image data augmentation generator
datagen = ImageDataGenerator(height_shift_range = 0.5)

# prepare iterator
it = datagen.flow(samples, batch_size = 1)

# generate samples and plot
fig = plt.figure(figsize = (20, 5))
for i in range(10):
  # define subplot
  plt.subplot(2, 5, i+1)
  # generate batch of images
  batch = it.next()
  # convert to unsigned integers for viewing
  image = batch[0].astype('uint8')
  # plot raw pixel data
  plt.imshow(image)
```

<h1><font color = 'brown' size = '6'>
<b>
Random Rotation Augmentation
</b>
</font>
</h1>

<h1>
<b>
<ul>
<font color = 'brown green' size = '5'>

<li>
A rotation augmentation randomly rotates the image clockwise by a given number of degrees from 0 to 360.
</li><br>

<li>
The rotation will likely rotate pixels out of the image frame and leave areas of the frame with no pixel data that must be filled in.
</li><br>

<li>
The example below demonstrates random rotations via the <i>rotation_range</i> argument, with rotations to the image between 0 and 90 degrees.
</li>

</font>
</ul>

</b>
</h1>

```python
# example of random rotation image augmentation
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bird.jpg')

# convert to numpy array
data = img_to_array(img)

# expand dimension to one sample
samples = np.expand_dims(data, 0)

# create image data augmentation generator
datagen = ImageDataGenerator(rotation_range = 90)

# prepare iterator
it = datagen.flow(samples, batch_size = 1)

# generate samples and plot
fig = plt.figure(figsize = (20, 5))
for i in range(10):
  # define subplot
  plt.subplot(2, 5, i+1)
  # generate batch of images
  batch = it.next()
  # convert to unsigned integers for viewing
  image = batch[0].astype('uint8')
  # plot raw pixel data
  plt.imshow(image)
```

<h1><font color = 'brown' size = '6'>
<b>
Random Brightness Augmentation
</b>
</font>
</h1>

<h1>
<b>
<ul>
<font color = 'brown green' size = '5'>

<li>
The brightness of the image can be augmented by either randomly darkening images, brightening images, or both.
</li><br>

<li>
The intent is to allow a model to generalize across images trained on different lighting levels.
</li><br>

<li>
This can be achieved by specifying the <i>brightness_range</i> argument to the <i>ImageDataGenerator()</i> constructor that specifies min and max range as a float representing a percentage for selecting a darkening or brightening amount.
</li><br>

<li>
Values less than 1.0 darken the image, e.g. [0.5, 1.0], whereas values larger than 1.0 brighten the image, e.g. [1.0, 1.5], where 1.0 has no effect on brightness.
</li><br>

<li>
The example below demonstrates a brightness image augmentation, allowing the generator to randomly darken the image between 1.0 (no change) and 0.2 or 20%.
</li>

</font>
</ul>

</b>
</h1>

```python
# example of brighting image augmentation
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bird.jpg')

# convert to numpy array
data = img_to_array(img)

# expand dimension to one sample
samples = np.expand_dims(data, 0)

# create image data augmentation generator
datagen = ImageDataGenerator(brightness_range = [0.2, 1.0])

# prepare iterator
it = datagen.flow(samples, batch_size = 1)

# generate samples and plot
fig = plt.figure(figsize = (20, 5))
for i in range(10):
  # define subplot
  plt.subplot(2, 5, i+1)
  # generate batch of images
  batch = it.next()
  # convert to unsigned integers for viewing
  image = batch[0].astype('uint8')
  # plot raw pixel data
  plt.imshow(image)
```

<h1><font color = 'brown' size = '6'>
<b>
Random Zoom Augmentation
</b>
</font>
</h1>

<h1>
<b>
<ul>
<font color = 'brown green' size = '5'>

<li>
A zoom augmentation randomly zooms the image and either adds new pixel values around the image or interpolates pixel values respectively.
</li><br>

<li>
Image zooming can be configured by the <i>zoom_range</i> argument to the <i>ImageDataGenerator</i> constructor.
</li><br>

<li>
You can specify the percentage of the zoom as a single float or a range as an array or tuple.
</li><br>

<li>
If a float is specified, then the range for the zoom will be [1-value, 1+value].
</li><br>

<li>
For example, if you specify 0.3, then the range will be [0.7, 1.3], or between 70% (zoom in) and 130% (zoom out).
</li><br>

<li>
The zoom amount is uniformly randomly sampled from the zoom region for each dimension (width, height) separately.
</li>

</font>
</ul>

</b>
</h1>

```python
# example of brighting image augmentation
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bird.jpg')

# convert to numpy array
data = img_to_array(img)

# expand dimension to one sample
samples = np.expand_dims(data, 0)

# create image data augmentation generator
datagen = ImageDataGenerator(zoom_range = [0.5, 1.0])

# prepare iterator
it = datagen.flow(samples, batch_size = 1)

# generate samples and plot
fig = plt.figure(figsize = (20, 5))
for i in range(10):
  # define subplot
  plt.subplot(2, 5, i+1)
  # generate batch of images
  batch = it.next()
  # convert to unsigned integers for viewing
  image = batch[0].astype('uint8')
  # plot raw pixel data
  plt.imshow(image)
```
